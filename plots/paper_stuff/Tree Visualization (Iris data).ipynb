# Tree Visualization (Iris data)

```python
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

```python
iris_data = load_iris()
```

```python
iris_data.keys()
```

```python
iris_df = {"x": iris_data["data"][:,[0,2]],
           "y": iris_data["target"],
           "x_names": iris_data["feature_names"],
           "y_names": iris_data["target_names"]}
```

```python
dt = DecisionTreeClassifier(max_depth=3)

dt.fit(X=iris_df["x"],
       y=iris_df["y"])
```

```python
%matplotlib inline

plt.figure(figsize=(30,18))
plot_tree(dt, feature_names=iris_data["feature_names"],
              class_names  =iris_data["target_names"],
         filled=False, rounded=False, fontsize=35);
plt.savefig('tree_viz.eps', format='eps')
```

```python
%matplotlib inline

plt.figure(figsize=(30,18))
plot_tree(dt, feature_names=iris_data["feature_names"],
              class_names  =iris_data["target_names"],
         filled=True, rounded=False, fontsize=35);
plt.savefig('tree_viz_col.eps', format='eps')
```
